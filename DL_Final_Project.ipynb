```python
import json
import logging
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs
```

```python
with open(r"dev-v2.0.json", "r") as read_file:
    train_data = json.load(read_file)

with open(r"dev-v1.1.json", "r") as read_file:
    test_data = json.load(read_file)
```

```python
train_data = [word for topic in train_data['data'] for word in topic['paragraphs']]
test_data = [word for topic in test_data['data'] for word in topic['paragraphs']]
```

```python
transformer_spec = {
    'learning_rate': 3e-5,
    'num_train_epochs': 2,
    'max_seq_length': 384,
    'doc_stride': 128,
    'train_batch_size': 2,
    'gradient_accumulation_steps': 8,
}
```

```python
transformer_qa = QuestionAnsweringModel("bert", "bert-base-cased", transformer_spec)   
```

```python
transformer_qa.train_model(train_data, eval_data = test_data)
```

```python
results, texts = transformer_qa.eval_model(test_data)
```
